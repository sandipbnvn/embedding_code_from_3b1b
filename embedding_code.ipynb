```python
import gensim.downloader
model = gensim.downloader.load('glove-wiki-gigaword-50')
model['tower']
```

```python
em_tower = model['tower']
em_dome = model['dome']
em_cat = model['cat']
em_dog = model['dog']

em_man = model['man']
em_woman = model['woman']
em_uncle = model['uncle']
em_aunt = model['aunt']
em_girl = model['girl']
em_boy = model['boy']

```

```python
import numpy as np
def cos_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)

    magnitude_vec1 = np.linalg.norm(vec1)
    magnitude_vec2 = np.linalg.norm(vec2)

    
    cosine_sim = dot_product / (magnitude_vec1 * magnitude_vec2)
    return np.round(cosine_sim, 2)
```

```python

print(f'cosine similiarity between dome and tower: {cos_similarity(em_dome, em_tower):.2f}')
print(f'cosine similiarity between dog and cat: {cos_similarity(em_dog, em_cat):.2f}')
```

```python
print(f'cosine similiarity between tower and cat: {cos_similarity(em_tower, em_cat):.2f}')
```

```python
em_gender1 = em_man - em_woman
em_gender2 = em_uncle - em_aunt
print(f'cosine similiarity between gender1 and gender2: {cos_similarity(em_gender1, em_gender2):.2f}')
```
